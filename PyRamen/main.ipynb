```python

### Import csv's for menu and sales data using pathlib
import csv
from pathlib import Path
import pprint
import sys

menu_filepath = Path('..\PyRamen\hw_02-menu_data.csv')
sales_filepath = Path('..\PyRamen\hw_02-sales_data.csv')
export_file = Path('../PyRamen/report.txt')


### Initialize list objects to menu and sales data

menu = []
sales = []

## Menu data separated for indexing and compiled indexed menu data

indexed_menu_data = [menu_items,
            menu_categories,
            menu_descriptions,
            menu_prices,
            menu_costs]

menu_items = []
menu_categories = []
menu_descriptions = []
menu_prices = []
menu_costs = []

### Create list of lists constituting the complete menu and lists of menu data separated by type of information for indexing, the latter of which which will make cross-referencing against sales data easier later on

with open(menu_filepath, "r") as menu_data:
    menu_reader = csv.reader(menu_data)
    header = next(menu_reader)

    for row in menu_reader:
        menu.append(row)
        menu_items.append(row[0])
        menu_categories.append(row[1])
        menu_descriptions.append(row[2])
        menu_prices.append(row[3])
        menu_costs.append(row[4])

### If you would like to cofirm that the separated menu data was transcribed correctly from the menu and was properly indexed, or in the event that data appears incorrect or to have been partially omitted between the menu and the menu data separated for indexing, uncomment and allow the code to run with the blocks beneath Issue 1 and/or Issue 2, respectively, active to isolate the issue. These lines have been commented out to reduce computation required for code to run, but if need be, the code can run with both issue-checks active.


    ## Issue 1: Determine incorrect data
    def Determine_Incorrectly_Transcribed_Data():
        for item in menu:
            # Initialize empty list to hold incorrect data
            Incorrect_Entries = []
            # For each entry in the menu, compare information in lists of entries constituting menu by index within those lists with information in separated lists constituting indexed menu data by the index of the information type and the index within the list of that information type corresponding to the index of the information in the menu (ensuring both the information and the index are congruent across the menu and separated menu data)
            if menu[menu.index(item)][0] == indexed_menu_data[0][menu.index(item)] and menu[menu.index(item)][1] == indexed_menu_data[1][menu.index(item)] and menu[menu.index(item)][2] == indexed_menu_data[2][menu.index(item)] and menu[menu.index(item)][3] == indexed_menu_data[3][menu.index(item)] and menu[menu.index(item)][4] == indexed_menu_data[4][menu.index(item)]:
                return None
            # If the two chunks of information are not congruent, show the pieces of menu data for that chunck side-by-side with pieces of data corresponding to the checked index within the separated lists of menu data for human review (if the pieces of information are in fact identical, the issue in transcribing from the menu was in properly indexing within the indexed menu data list)
            else:
                Issue_Entry = (f"\n\n             Menu Information | Separated Information\n"
                    f"\n Item Name:             {menu[menu.index(item)][0]} | {menu_items[menu.index(item)]}\n" 
                    f" Item Category:         {menu[menu.index(item)][1]} | {menu_categories[menu.index(item)]}\n"
                    f" Item Description:      {menu[menu.index(item)][2]} | {menu_descriptions[menu.index(item)]}\n"
                    f" Item Price:            {menu[menu.index(item)][3]} | {menu_prices[menu.index(item)]}\n" 
                    f" Item Cost:             {menu[menu.index(item)][4]} | {menu_costs[menu.index(item)]}\n")
                Incorrect_Entries.append(Issue_Entry)
                print(f"The following entry data is incongruent: {Issue_Entry}")

        if len(Incorrect_Entries) > 0:
            sys.exit()   

    Determine_Incorrectly_Transcribed_Data()

    ## Issue 2: Determine if data ommitted
    for separated_data_list in indexed_menu_data:
        if len(menu) == len(separated_data_list):
            None
        elif len(menu) > len(separared_data_list):
            print(f"Menu has too many entries or {separared_data_list} is missing entries")
        else:
            print(f"{separared_data_list} has too many entries or menu is missing entries")

    # Citation for f-string new lines: https://stackoverflow.com/questions/11497376/how-do-i-specify-new-lines-on-python-when-writing-on-files
    # Citation for for sys.exit(): https://www.hashbangcode.com/article/stopping-code-execution-python#:~:text=To%20stop%20code%20execution%20in,Here%20is%20a%20simple%20example.
        


### Create list of lists for sales
with open(sales_filepath, "r") as sales_data:
    sales_reader = csv.reader(sales_data)
    header = next(sales_reader)

    for row in sales_reader:
        sales.append(row)

### Initialize empty dictionary to hold report
    report = {}

### Initialize a row counting variable
    row_count = 0

### Loop over every row in sales and populate dictionary using information from indexed menu data
    for sale in sales:   
        row_count += 1

        ## Empty list to hold erroneous sales data, should there be any
        erroroneous_sales_items = []

        ## Specify location of sales item and quantity in list sales items
        sales_item = sale[4] 
        Quantity = int(sale[3])
        
        ## Ensure every item passed from sales into report is on the menu (to account for human error in sales data entry or discontinuation of an item)
        if sales_item not in menu_items:
            erroroneous_sales_items.append(sales_item)
        
        else: 
            ## Create menu item index variable to cross-reference
            Menu_Item = indexed_menu_data[0][menu_items.index(sales_item)]
            ## Indicate indexed points for price and cost in menu data against which to cross-reference each sales item for profit etc. calculations
            price = float(indexed_menu_data[3][indexed_menu_data[0].index(Menu_Item)])
            cost = float(indexed_menu_data[4][indexed_menu_data[0].index(Menu_Item)])
            profit = price - cost

            ## Determine whether a given item is already in the report, if so adding to its metrics totals
            if  sales_item in report.keys(): # Citation: https://www.geeksforgeeks.org/python-check-whether-given-key-already-exists-in-a-dictionary/
                report[sales_item]["01-count"] += Quantity
                report[sales_item]["02-revenue"] += price * Quantity
                report[sales_item]["03-cogs"] += cost * Quantity
                report[sales_item]["04-profit"] += profit * Quantity

            ## If not in the report, create a key-value pair with metrics to track
            else:
                report[sales_item]={"01-count": Quantity,
                                "02-revenue": price * Quantity,
                                "03-cogs": cost * Quantity,
                                "04-profit": profit * Quantity}

        ## Identify erroneous sales items within report, if there are any
        if len(erroroneous_sales_items) > 0:
            report['THE FOLLOWING ITEMS WERE IN SALES RECORD, BUT ARE NOT ON THE MENU]'] = erroroneous_sales_items
        else:
            None
            
    ### Export file
    # Citation for formatted dictionary print: https://stackoverflow.com/questions/44689546/how-to-print-out-a-dictionary-nicely-in-python/44689627  
    Report = pprint.pformat(report)
    print(Report)

    with open(export_file, "w") as text_file:
        text_file.write (Report)

```
