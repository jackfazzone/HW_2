```python

# Import csv's for menu and sales data using pathlib
import csv
from pathlib import Path
import pprint

menu_filepath = Path('..\PyRamen\hw_02-menu_data.csv')
sales_filepath = Path('..\PyRamen\hw_02-sales_data.csv')
export_file = Path('../PyRamen/report.txt')


# Initialize list objects to menu and sales data

menu = []
sales = []

# Menu data
menu_items = []
menu_categories = []
menu_descriptions = []
menu_prices = []
menu_costs = []
erroroneous_sales_items = []

# Create list of lists for menu and list of indexed data for menu data

with open(menu_filepath, "r") as menu_data:
    menu_reader = csv.reader(menu_data)
    header = next(menu_reader)

    for row in menu_reader:
        menu.append(row)
        menu_items.append(row[0])
        menu_categories.append(row[1])
        menu_descriptions.append(row[2])
        menu_prices.append(row[3])
        menu_costs.append(row[4])


# Create list of lists for sales
with open(sales_filepath, "r") as sales_data:
    sales_reader = csv.reader(sales_data)
    header = next(sales_reader)

    for row in sales_reader:
        sales.append(row)


# Initialize empty dictionary to hold report
    report = {}

# Initialize a row counting variable
    row_count = 0

# Loop over every row in sales and populate dictionary using information from indexed menu data
    for sale in sales:   
        row_count += 1
        # Specify location of sales item and quantity in list sales items
        sales_item = sale[4] 
        Quantity = int(sale[3])
        
        
        # Ensure every item passed from sales into report is on the menu (to account for human error in sales data entry or discontinuation of an item)
        if sales_item not in menu_items:
            report[-1] = {'THE FOLLOWING ITEMS WERE IN SALES RECORD, BUT ARE NOT ON THE MENU' : erroroneous_sales_items}
            erroroneous_sales_items.append(sales_item)
        
        else: 
            # Create menu item index variable to cross-reference
            Menu_Item = menu_items.index(sales_item)
            # Indicate indexed points for price and cost in menu data against which to cross-reference each sales item for profit etc. calculations
            price = float(menu_prices[Menu_Item])
            cost = float(menu_costs[Menu_Item])
            profit = price - cost

            # Determine whether a given item is already in the report, if so adding to its metrics totals
            if  sales_item in report.keys(): # Citation: https://www.geeksforgeeks.org/python-check-whether-given-key-already-exists-in-a-dictionary/
                report[sales_item]["01-count"] += Quantity
                report[sales_item]["02-revenue"] += price * Quantity
                report[sales_item]["03-cogs"] += cost * Quantity
                report[sales_item]["04-profit"] += profit * Quantity

            # If not in the report, create a key-value pair with metrics to track
            else:
                report[sales_item]={"01-count": Quantity,
                                "02-revenue": price * Quantity,
                                "03-cogs": cost * Quantity,
                                "04-profit": profit * Quantity}
            
    #export file
    # Citation: https://stackoverflow.com/questions/44689546/how-to-print-out-a-dictionary-nicely-in-python/44689627  
    Report = pprint.pformat(report)
    print(Report)

    with open(export_file, "w") as text_file:
        text_file.write (Report)

```
